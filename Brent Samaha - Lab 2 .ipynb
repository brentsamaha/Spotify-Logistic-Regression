## Question 4. Build a logistic regression model

```python
#Dropping non numeric variables
train_data = train_data.drop(['track', 'artist', 'uri'], axis=1)
val_data = val_data.drop(['track', 'artist', 'uri'], axis=1)
```

```python
#Removing target variable 
x_train = train_data.iloc[:,:-1]
x_val = val_data.iloc[:, :-1]
```

```python
x_train.dtypes
```

```python
for field in x_train.columns:
    standard_dev = x_train[field].std()
    mean = x_train[field].mean()
    x_train[field] = (x_train[field] - mean) / standard_dev
    x_val[field] = (x_val[field] - mean) / standard_dev
```

```python
log_model = LogisticRegression(random_state=13).fit(x_train, train_data.target)
```

```python
log_model.predict_proba(x_train)
```

```python
#Getting the target predictions to test against our train_data.target
y_train_pred = log_model.predict(x_train)
y_train_pred
```

```python
#Getting target predictions to test against our val_data.target
y_val_pred = log_model.predict(x_val)
y_val_pred
```

### Question 4. a)

```python
#Accuracy:

train_acc = metrics.accuracy_score(train_data.target, y_train_pred)
val_acc = metrics.accuracy_score(val_data.target, y_val_pred)

print("The training dataset accuracy is ",train_acc, "\n","The validation accuracy is ", val_acc, sep="")
```

```python
#Precision

train_prec = metrics.precision_score(train_data.target, y_train_pred)
val_prec = metrics.precision_score(val_data.target, y_val_pred)

print("The training dataset precision is ",train_prec, "\n","The validation precision is ", val_prec, sep="")
```

```python
#Recall

train_rec = metrics.recall_score(train_data.target, y_train_pred)
val_rec = metrics.recall_score(val_data.target, y_val_pred)

print("The training dataset recall is " , train_rec, "\n","The validation recall is " , val_rec, sep="")
```

```python
#AUC:
```

```python
y_train_predprob = log_model.predict_proba(x_train)
y_val_predprob = log_model.predict_proba(x_val)
```

```python
fpr, tpr, thresholds = metrics.roc_curve(train_data.target, y_train_predprob[:,1], pos_label = 1)
```

```python
train_auc = metrics.auc(fpr, tpr)
```

```python
print("The training dataset auc is " , train_auc)
```

```python
fpr, tpr, thresholds = metrics.roc_curve(val_data.target, y_val_predprob[:,1], pos_label = 1)
val_auc = metrics.auc(fpr, tpr)
```

```python
print("The validation AUC is " , val_auc)
```

```python
#Being above .5, the AUC details that the model is better than classifying at random
```

## Question 5. Feature engineering - variable transformation 

```python
#duration until chorus as per song length

x_train["chorus_hit_duration_interaction"] = x_train.chorus_hit  / x_train.duration_ms
x_val["chorus_hit_duration_interaction"] = x_val.chorus_hit / x_val.duration_ms
```

#### I chose to create this interaction with the thinking that a hit would more likely have their chorus earlier in a song. My thinking is that hits are generally shorter songs, so if a long song is a hit, maybe it's because its chorus comes earlier in the song. By starting earlier in a long song, a listener could be more inclinced to listen to the whole song, like it more, making it a hit. Maybe! 

```python
x_train['dance_valence'] = (x_train.danceability * x_train.valence)
x_val['dance_valence'] = (x_val.danceability * x_val.valence)
```

#### I chose to multiply the values of danceability and valence as they had one of the stronger instances of correlation in the training dataset. 

```python
x_train.loudness.agg(['skew']).transpose()
```

```python
x_train["loudness_sqr"] = (x_train["loudness"] * x_train["loudness"])
x_val["loudness_sqr"] = (x_val["loudness"] * x_val["loudness"])
```

#### I chose to square the value of loudness as I looked to the skewd-ness of the value in the training dataset and saw that it was negatively skewed. I believe that by squaring it, there could be some effect to make the distribution more normal.
